```python
def rfm_seg(df_fitted):
    if (df_fitted['recency_scores'] in range(4,6) or df_fitted['frequency_scores'] in range(4,6) or
        df_fitted['monetary_scores'] in range (4,6)):
        return  'Champions' #Champions: RFM score range - R: 4-5, F: 4-5, M: 4-5
    
    elif (df_fitted['recency_scores'] in range(3,5) or df_fitted['frequency_scores'] in range(3,5) or
        df_fitted['monetary_scores'] in range (3,5)):
        return 'Loyal customers' #Loyal customers: RFM score range - R: 3-5, F: 3-5, M: 3-5
    
    elif (df_fitted['recency_scores'] in range(4,5) or df_fitted['frequency_scores'] in range(2,3) or
        df_fitted['monetary_scores'] in range (2,3)):
        return 'Potential loyalist' #Potential loyalist: RFM score range - R: 4-5, F: 2-3, M: 2-3
    
    elif (df_fitted['recency_scores'] in range(4,5) or df_fitted['frequency_scores'] in range(1,2) or
        df_fitted['monetary_scores'] in range (1,2)):
        return 'Recent customers' #Recent customers: RFM score range - R: 4-5, F: 1-2, M: 1-2
    
    elif (df_fitted['recency_scores'] in range(4,5) or df_fitted['frequency_scores'] in range(1,2) or
        df_fitted['monetary_scores'] in range (1,2)):
        return 'Promising' #Promising: RFM score range - R: 4-5, F: 1-2, M: 1-2
    
    elif (df_fitted['recency_scores'] in range(3,5) or df_fitted['frequency_scores'] in range(3,5) or
        df_fitted['monetary_scores'] in range (3,5)):
        return 'Needs attention' #Needs attention: RFM score range - R: 3-5, F: 3-5, M: 3-5
    
    elif (df_fitted['recency_scores'] in range(1,2) or df_fitted['frequency_scores'] in range(1,2) or
        df_fitted['monetary_scores'] in range (1,2)):
        return 'About to sleep' #About to sleep: RFM score range - R: 1-2, F: 1-2, M: 1-2
    
    elif (df_fitted['recency_scores'] in range(2,5) or df_fitted['frequency_scores'] in range(1,3) or
        df_fitted['monetary_scores'] in range (1,3)):
        return 'At risk' #At risk: RFM score range - R: 2-5, F: 1-3, M: 1-3
    
    elif (df_fitted['recency_scores'] in range(1,3) or df_fitted['frequency_scores'] in range(4,5) or
        df_fitted['monetary_scores'] in range (4,5)):
        return "Can't lose them" #Can't lose them: RFM score range - R: 1-3, F: 4-5, M: 4-5
    
    elif (df_fitted['recency_scores'] in range(1,2) or df_fitted['frequency_scores'] in range(1,2) or
        df_fitted['monetary_scores'] in range (1,2)):
        return 'Hibernating' #Hibernating: RFM score range - R: 1-2, F: 1-2, M: 1-2
    else:
        return 'Hibernating'
    
df_fitted['RFM_Label'] = df_fitted.apply(rfm_seg, axis=1)
df_fitted.head(20)
```

Findings:

The RFM analysis identified 10 distinct customer segments based on their recency, frequency, and monetary value.
The "Champions" segment is the most valuable, with high recency, frequency, and monetary value.
The "Loyal Customers" and "Potential Loyalist" segments are also valuable, as they have high frequency and monetary value.
The "Recent Customers" and "Promising" segments have high recency but lower frequency and monetary value. These customers may be at risk of churn if they are not engaged with.
The "Needs Attention" and "About to Sleep" segments have low recency, frequency, and monetary value. These customers may be at risk of churn if they are not engaged with.
The "At Risk" and "Can't Lose Them" segments have high monetary value but low recency and frequency. These customers may be at risk of churn if they are not engaged with.
The "Hibernating" segment has low recency, frequency, and monetary value. These customers are likely to be lost if they are not engaged with.
Strategies:

Champions: Continue to provide excellent service to these customers and offer them exclusive benefits and rewards.
Loyal Customers and Potential Loyalists: Focus on increasing their frequency of purchase and monetary value. This can be done by offering them personalized discounts, promotions, and loyalty programs.
Recent Customers and Promising: Engage with these customers to increase their frequency of purchase and monetary value. This can be done by sending them targeted emails, offering them special promotions, and providing excellent customer service.
Needs Attention and About to Sleep: Develop strategies to re-engage these customers. This can be done by sending them personalized emails, offering them special promotions, and providing excellent customer service.
At Risk and Can't Lose Them: Focus on increasing their frequency of purchase and monetary value. This can be done by offering them personalized discounts, promotions, and loyalty programs.
Hibernating: These customers are likely to be lost if they are not engaged with. Develop strategies to re-engage these customers. This can be done by sending them personalized emails, offering them special promotions, and providing excellent customer service.
By implementing these strategies, businesses can improve customer retention and increase revenue.
